```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(start=-5, stop=5, step=0.5)
x
```

```python
def model(x:float, w:float = 1, b:float = 0):
    return x*w +b
```

```python
plt.figure(figsize=(10, 7))
plt.plot(x, model(x))
plt.title("No bias")
plt.xlabel("Input")
plt.ylabel("Preds")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.plot(x, model(x, b=2))
plt.title("bias = 2")
plt.xlabel("Input")
plt.ylabel("Preds")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.plot(x, model(x, b=2, w=2))
plt.title("Weight and bias = 2")
plt.xlabel("Input")
plt.ylabel("Preds")
plt.grid(True)
plt.show()
```
